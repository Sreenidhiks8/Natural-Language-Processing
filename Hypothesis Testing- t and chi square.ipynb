```python
import nltk
f = open('article.txt', encoding="utf8")
raw = f.read()
words = nltk.word_tokenize(raw)
words=[word.lower() for word in words if word.isalpha()]
print(words)
count1=0
count2=0

count3=0
count4=0

count5=0
count6=0

#to find collocation for speech & recognition
for i in words:
    if i.lower()=='speech':
        count1=count1+1

for i in words:
    if i.lower()=='recognition':
        count2=count2+1
print()

print("No of occurrences of Speech: ", count1)
print()
print("No of occurrences of Recognition : ", count2)
print()
print("Tot words in corpus are: ")
print(len(words))
print()

#to find collocation for spam & detection
for i in words:
    if i.lower()=='spam':
        count3=count3+1

for i in words:
    if i.lower()=='detection':
        count4=count4+1
print()

print("No of occurrences of Spam: ", count3)
print()
print("No of occurrences of Detection : ", count4)
print()
print("Tot words in corpus are: ")
print(len(words))
print()

#to find collocation for machine & translation

for i in words:
    if i.lower()=='machine':
        count5=count5+1

for i in words:
    if i.lower()=='translation':
        count6=count6+1
print()

print("No of occurrences of Machine: ", count5)
print()
print("No of occurrences of Translation : ", count6)
print()
print("Tot words in corpus are: ")
print(len(words))
print()
```

```python
#T test for speech and recognition
from math import sqrt

p_speech = 8/842
print(p_speech)

p_recognition= 6/842
print(p_recognition)

p_speech_recognition= p_speech*p_recognition    
print(p_speech_recognition,"\n")    

xbar=5/842
print("xbar= ",xbar)     
print()

s2=xbar*(1-xbar)
print("s2= ",s2)
print()

t=(xbar- p_speech_recognition)/sqrt((s2/842))
print("t value= ",t)      
print()       

critical_val=2.576  #for t test

if (t<critical_val): 
    print("Speech Recognition is not a collocation")
else:
    print("Speech Recognition is not a collocation")
```

```python
#T test for spam & detection
from math import sqrt

p_spam = 5/842
print(p_spam)

p_detection= 4/842
print(p_detection)

p_spam_detection= p_spam*p_detection    
print(p_spam_detection,"\n")    

xbar=4/842
print("xbar= ",xbar)     
print()

s2=xbar*(1-xbar)
print("s2= ",s2)
print()

t=(xbar- p_spam_detection)/sqrt((s2/842))
print("t value= ",t)      
print()       

critical_val=2.576  #for t test

if (t<critical_val): 
    print("Spam Detection is not a collocation")
else:
    print("Spam Detection is not a collocation")
```

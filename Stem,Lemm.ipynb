```python
import nltk
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import webtext
text= webtext.raw()
sent_tokenizer=PunktSentenceTokenizer(text)
sents_1= sent_tokenizer.tokenize(text)
print(sents_1[0])
```

```python
from nltk.corpus import wordnet as wn
import nltk
nltk.download('wordnet')
import nltk
nltk.download('omw-1.4')
syn=wn.synsets('dog')[0]
syn.name()
```

```python
syn.definition()
```

```python
syn.examples()
syn.hypernyms()  #more abstract terms
syn.hyponyms()    #more similar terms
```

```python
syn.hypernyms()[0].hyponyms()
```

```python
syn.root_hypernyms()
```

```python
lemmas= syn.lemmas()
len(lemmas)
```

```python
for i in lemmas:
    print(i.name())
```

```python
syn1=wn.synset('good.n.02')
lemmas1= syn1.lemmas()
for i in lemmas1:
    print(i.antonyms())
```

```python
syn1=wn.synset('good.a.01')
lemmas1=syn1.lemmas()
for i in lemmas1:
    for j in i.antonyms():
        print(j.name())
```
